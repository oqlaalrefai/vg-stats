#### the median of North American video game sales

```python
df[['Name', 'NA_Sales']].median()[0]
df[df['NA_Sales'] == df[['Name', 'NA_Sales']].median()[0]].head()
```

```python
x = (df['NA_Sales'].head(1))
meanValue = df["NA_Sales"].mean()
standardDeviation = (x - meanValue)/df["NA_Sales"].std()
standardDeviation[0]
```
